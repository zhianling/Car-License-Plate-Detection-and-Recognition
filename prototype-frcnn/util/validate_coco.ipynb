```python
import json
import os
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

def visualize_coco_annotation(image_path, coco_json_path):
    """
    Visualizes COCO JSON annotations on a given image.

    Args:
        image_path (str): Path to the image file.
        coco_json_path (str): Path to the COCO JSON annotation file.
    """

    # Load the COCO JSON data
    with open(coco_json_path, 'r') as f:
        coco_data = json.load(f)

    # Load the image
    try:
        img = Image.open(image_path)
    except FileNotFoundError:
        print(f"Error: Image file not found at {image_path}")
        return
    except Exception as e:
        print(f"Error loading image: {e}")
        return

    # Create a drawing context
    draw = ImageDraw.Draw(img)

    # Find the image ID in the JSON data (assuming only one image in JSON)
    if not coco_data['images']:
        print("Error: No image information found in COCO JSON.")
        return
    image_id = coco_data['images'][0]['id']

    # Get category names (assuming you have a 'categories' field)
    category_names = {}
    if 'categories' in coco_data:
        for category in coco_data['categories']:
            category_names[category['id']] = category['name']
    else:
        print("Warning: No 'categories' field found in COCO JSON. Using category IDs as labels.")

    # Iterate through annotations and draw bounding boxes
    for ann in coco_data['annotations']:
        if ann['image_id'] == image_id:
            bbox = ann['bbox']  # [x, y, width, height]
            category_id = ann['category_id']

            # Convert COCO bbox to (x1, y1, x2, y2) for drawing
            x1 = bbox[0]
            y1 = bbox[1]
            x2 = x1 + bbox[2]
            y2 = y1 + bbox[3]

            # Draw the rectangle
            draw.rectangle([x1, y1, x2, y2], outline="red", width=2)

            # Add label (if available)
            if category_id in category_names:
                label = category_names[category_id]
            else:
                label = str(category_id)  # Use ID if name not found
            draw.text((x1, y1 - 10), label, fill="red")

    # Display the image using matplotlib (for Jupyter Notebook)
    plt.imshow(img)
    plt.axis('off')  # Turn off axis labels
    plt.show()

    # Optional: Save the image
    # img.save("annotated_image.jpg")


if __name__ == '__main__':
    # --- Example Usage ---
    # Replace with the *absolute* paths to your image and COCO JSON file
    image_path = "../data/frame/video_001/frame00100.jpg"  # Example path
    coco_json_path = "../data/annotation/coco/annotation_cvat_001/annotation_cvat_001_coco_frame00100.json"  # Example path

    visualize_coco_annotation(image_path, coco_json_path)
```
